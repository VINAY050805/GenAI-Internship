```python
import pandas as pd
import sqlite3
```

```python
orders_df = pd.read_csv("orders.csv")
orders_df.head()
```

```python
users_df = pd.read_json("users.json")
users_df.head()
```

```python
conn = sqlite3.connect(":memory:")
cursor = conn.cursor()

with open("restaurants.sql", "r") as file:
    sql_script = file.read()

cursor.executescript(sql_script)

restaurants_df = pd.read_sql(
    "SELECT * FROM restaurants",
    conn
)

restaurants_df.head()
```

```python
orders_users_df = pd.merge(
    orders_df,
    users_df,
    on="user_id",
    how="left"
)

orders_users_df.head()
```

```python
final_df = pd.merge(
    orders_users_df,
    restaurants_df,
    on="restaurant_id",
    how="left"
)

final_df.head()
```

```python
final_df = pd.merge(
    orders_users_df,
    restaurants_df,
    on="restaurant_id",
    how="left"
)

final_df.head()
```

```python
final_df.to_csv("final_food_delivery_dataset.csv", index=False)
print(" final_food_delivery_dataset.csv created")
```

```python
print("Orders rows:", orders_df.shape[0])
print("Final dataset rows:", final_df.shape[0])
print("Distinct users:", final_df["user_id"].nunique())
```

```python
final_df["order_date"] = pd.to_datetime(
    final_df["order_date"],
    dayfirst=True
)

final_df["quarter"] = final_df["order_date"].dt.to_period("Q")
```

```python
final_df["rating_range"] = pd.cut(
    final_df["rating"],
    bins=[3.0, 3.5, 4.0, 4.5, 5.0],
    labels=["3.03.5", "3.64.0", "4.14.5", "4.65.0"],
    include_lowest=True
)

final_df.groupby("rating_range")["total_amount"].sum().sort_values(ascending=False)
```

```python
restaurant_count = final_df.groupby("cuisine")["restaurant_id"].nunique()
revenue_by_cuisine = final_df.groupby("cuisine")["total_amount"].sum()

pd.concat(
    [restaurant_count, revenue_by_cuisine],
    axis=1,
    keys=["restaurant_count", "total_revenue"]
)
```
