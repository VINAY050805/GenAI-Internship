# 1Ô∏è‚É£ Unique Words in a Sentence

```python
# Problem 1: Unique Words in a Sentence

sentence = input("Enter sentence: ")

words = sentence.split()     # Split sentence into words
unique_words = set(words)    # Remove duplicates using set

print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

# 2Ô∏è‚É£ Highest Salary from Employee Data

```python
# Problem 2: Highest Salary

employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

highest_employee = max(employees, key=employees.get)
highest_salary = employees[highest_employee]

print("\nHighest Salary:", highest_employee, "-", highest_salary)
```

# 3Ô∏è‚É£ Find Maximum and Minimum Values

```python
# Problem 3: Max and Min from List

numbers = [45, 22, 89, 10, 66]

max_val = numbers[0]
min_val = numbers[0]

for num in numbers:
    if num > max_val:
        max_val = num
    if num < min_val:
        min_val = num

print("\nList:", numbers)
print("Max:", max_val)
print("Min:", min_val)
```

# 4Ô∏è‚É£ Count Products Above Price Threshold

```python
# Problem 4: Products above 1000

prices = [450, 1200, 899, 1500, 300]

count = 0
for price in prices:
    if price > 1000:
        count += 1

print("\nProducts above 1000:", count)
```

# 5Ô∏è‚É£ Calculate Attendance Percentage

```python
# Problem 5: Attendance Percentage

attendance = ["P", "P", "A", "P", "P"]

present_days = attendance.count("P")
total_days = len(attendance)

percentage = (present_days / total_days) * 100

print("\nAttendance Percentage:", percentage)
```

# 7Ô∏è‚É£ Count Character Frequency

```python
text = "python"
freq = {}

for char in text:
    if char in freq:
        freq[char] += 1
    else:
        freq[char] = 1

print("\nCharacter Frequency:", freq)
```
