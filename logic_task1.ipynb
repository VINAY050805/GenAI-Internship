# 1️⃣ User Login Check

```python
# Task 1: User Login Check

# Given credentials
username = "admin"
password = "1234"

# User input (for real-world simulation)
input_username = input("Enter username: ")
input_password = input("Enter password: ")

# Checking login
if input_username == username and input_password == password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

# 2️⃣ Pass / Fail Analyzer

```python
# Task 2: Pass / Fail Analyzer

marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0

for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

print("\nResult Analysis")
print("Passed Students:", pass_count)
print("Failed Students:", fail_count)
```

# 3️⃣ Simple Data Cleaner

```python
# Task 3: Simple Data Cleaner

names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = []

for name in names:
    cleaned_name = name.strip().lower()
    cleaned_names.append(cleaned_name)

print("\nCleaned Names:", cleaned_names)
```

# 4️⃣ Message Length Analyzer

```python
# Task 4: Message Length Analyzer

messages = ["Hi", "Welcome to the platform", "OK"]

print("\nMessage Analysis")

for msg in messages:
    length = len(msg)
    print(f"Message: '{msg}' | Length: {length}")

    if length > 10:
        print("This message is longer than 10 characters")
```

# 5️⃣ Error Message Detector

```python
# Task 5: Error Message Detector

logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = logs.count("ERROR")

print("\nTotal ERROR messages:", error_count)
```
